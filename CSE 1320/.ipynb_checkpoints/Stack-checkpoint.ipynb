# Stack

* linear data structure
* Opeartions are performed Last in First Out
  * LIFO

## Operations on a Stack

| Term    | Description                                |
|---------|--------------------------------------------|
| Push    | Adds an item to the stack                  |
| Pop     | Removes an item from the stack             |
| Peek    | Returns top element of stack               |
| IsEmpty | Returns TRUE if stack is empty, else FALSE |

```c
/* stack Demo */
#include <stdio.h>
#include <stdlib.h>
 
// A structure to represent a stack node
struct node
{
    int node_number;
    struct node *next_ptr;
};
 
int isEmpty(struct node *StackTop)
{
	if (StackTop == NULL)
		return 1;
	else
		return 0;
}
 
void push(struct node **StackTop, int NodeNumber)
{
    // Reserve memory for the data structure
    struct node *NewNode = malloc(sizeof(struct node));
    
    // Initialize values
    NewNode->node_number = NodeNumber;
	NewNode->next_ptr = NULL;
	
    // If the stack is empty
	if (*StackTop == NULL)
	{
		*StackTop = NewNode;
	}
	else
	{
        // Make the new node point to the old node
		NewNode->next_ptr = *StackTop;
        // Make the new node the StackTop
		*StackTop = NewNode;
    }

	printf("Stack node %d pushed with address %p and is pointing to %p\n",
		   (*StackTop)->node_number, *StackTop, (*StackTop)->next_ptr);	
	
}
 
void pop(struct node **StackTop)
{
    // Check if Stack is empty
	if (*StackTop == NULL)
	{
		printf("Pop not executed - stack is empty\n\n");
        return;
	}
	
    // Save StackTop
    struct node *TempPtr = *StackTop;

	printf("Stack node %d popped from address %p that was pointing to %p\n",
		   (*StackTop)->node_number, *StackTop, (*StackTop)->next_ptr);	

    free(*StackTop);
    // The new StackTop is the old node
    *StackTop = TempPtr->next_ptr;

	return;
 }
 
// Literally Print the StackTop
void peek(struct node *StackTop)
{
    if (isEmpty(StackTop))
        printf("The stack is empty\n\n");
    else
		printf("The top node has a value of %d at address %p and is pointing to %p\n\n",
		        StackTop->node_number, &*StackTop, StackTop->next_ptr);
}

void DisplayStack(struct node *StackTop)
{
	struct node *TempPtr = StackTop;
	
	if (StackTop == NULL)
		printf("\nStack is empty");
	
	while (TempPtr != NULL)
	{
		printf("Stack node %d is at address %p and is pointing to %p\n",
		        TempPtr->node_number, &*TempPtr, TempPtr->next_ptr);
        // We can do this recursively too
		TempPtr =  TempPtr->next_ptr;
	}
	
	return;
}
 
int main(void)
{
	struct node *StackTop = NULL;
	
    return 0;
}
```
