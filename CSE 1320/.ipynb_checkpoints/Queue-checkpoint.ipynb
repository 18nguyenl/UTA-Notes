# Queue

* linear data structure
* Operations performed First in First Out
  * FIFO
* It's like waiting in a line for food
  * first person in line gets what they want then next person goes
  * people line up at the back of the line

## Operations on a Queue

| Term    | Description                    |
|---------|--------------------------------|
| Enqueue | Adds an item to the queue      |
| Dequeue | Removes an item from the queue |
| Head    | Get the head of the queue      |
| Tail    | Get the tail of the queue      |

```c
/* queue Demo */
#include <stdlib.h>
#include <stdio.h>

#define HEAD 1
#define TAIL 0
 
// A structure to represent a queue node
struct node
{
    int node_number;
    struct node *next_ptr;
};

// Get in line
void enQueue(int NewNodeNumber, struct node **QueueHead, struct node **QueueTail)
{
    // Reserve Memory
    struct node *TempPtr = malloc(sizeof(struct node));
    
    // Initialize Values
    TempPtr->node_number = NewNodeNumber;
    TempPtr->next_ptr = NULL;

	/* Queue is empty */
    // Initialize the Queue
	if (*QueueHead == NULL)
	{
        // Only 1 element so QueueHead == QueueTail = TempPtr
		*QueueHead = *QueueTail = TempPtr;
	}
	else
	{
        // Get the node of the tail and point to the new node
		(*QueueTail)->next_ptr = TempPtr;
        // Make the new node the QueueTail
		*QueueTail = TempPtr;
	}
}
 
// Get out of line
void deQueue(struct node **QueueHead)
{
	struct node *TempPtr;
	
    // If queue is empty, return 0.
    if (*QueueHead == NULL)
	{
		printf("Queue is emtpy\n\n");
		return;
	}
	
	printf("Queue node %d is dequeuing\n", (*QueueHead)->node_number);
    // Store the next pointer
    // Get next person in line
    TempPtr = (*QueueHead)->next_ptr;
	free(*QueueHead);
    // Next person in line is now at the front
	*QueueHead = TempPtr;
}
 
void DisplayQueue(struct node *QueueHead)
{
	struct node *TempPtr = QueueHead;
	
    // Classic traversing
	while (TempPtr != NULL)
	{
		printf("Queue node %d is at address %p and is pointing to %p\n",
		        TempPtr->node_number, &*TempPtr, TempPtr->next_ptr);
		TempPtr =  TempPtr->next_ptr;
	}
	
	return;
} 

int isEmpty(struct node *QueueHead)
{
	if (QueueHead == NULL)
		return 1;
	else
		return 0;
}

void whosinline(int headortail, struct node *QueueHead, struct node *QueueTail)
{
    if (isEmpty(QueueHead))
	{
		printf("The queue is empty\n\n");
		return;
	}
    else if (headortail == HEAD)
		printf("The first node is %d at address %p and is pointing to %p\n\n",
		        QueueHead->node_number, QueueHead, QueueHead->next_ptr);
	else if (headortail == TAIL)
		printf("The last node is %d at address %p and is pointing to %p\n\n",
		        QueueTail->node_number, QueueTail, QueueTail->next_ptr);
}
 
int main(void)
{
	struct node *QueueHead, *QueueTail;

	system("clear");
	QueueHead = QueueTail = NULL;
	
	return 0;
}
```
